### Bus Lines Delay
### Visualization Plotting

```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ks_2samp
%matplotlib inline

path = 'Output_Data/'
os.listdir(path)
```

```python
week_game = pd.read_csv(path + 'BDM_Game_Weekday_Output.csv')
week_game_mean = week_game.groupby(['bus', 'interval'])['avg(duration)'].mean().to_frame().reset_index()

week_nogame = pd.read_csv(path + 'BDM_NoGame_Weekday_Output.csv')
week_nogame_mean = week_nogame.groupby(['bus', 'interval'])['avg(duration)'].mean().to_frame().reset_index()

weekend_game = pd.read_csv(path + 'BDM_Game_Weekend_Output.csv')
weekend_game_mean = weekend_game.groupby(['bus', 'interval'])['avg(duration)'].mean().to_frame().reset_index()

weekend_nogame = pd.read_csv(path + 'BDM_NoGame_Weekend_Output.csv')
weekend_nogame_mean = weekend_nogame.groupby(['bus', 'interval'])['avg(duration)'].mean().to_frame().reset_index()
weekend_game_mean.shape
```

```python
buses = ['BX1', 'BX13', 'BX6', 'Q48']

fig, (ax1, ax2, ax3, ax4) = plt.subplots(nrows = 4, ncols = 1, figsize = (16, 24))
fig.tight_layout(h_pad=8, w_pad=0)

for bus, ax in zip(buses, fig.axes):
    game = weekend_game_mean[weekend_game_mean['bus'] == bus]
    no_game = weekend_nogame_mean[weekend_nogame_mean['bus'] == bus]
    
    if len(game) != len(no_game):
        game = game[:-1]
        
    print ('{}\nGameday: Mean Trip Time ({}), Trip Time Std. Dev ({})'
           .format(bus, int(game['avg(duration)'].mean()), int(game['avg(duration)'].std())) +
           '\nNon-Gameday: Mean Trip Time ({}), Trip Time Std. Dev ({})'
           .format(int(no_game['avg(duration)'].mean()), int(no_game['avg(duration)'].std())) +
           '\nKS-Test: P-value ({})\n'.format(ks_2samp(game['avg(duration)'], no_game['avg(duration)'])[1]))

    width = 0.40
    ax.bar(range(len(game)), game['avg(duration)'], width,
           color='r', alpha=0.4, label='Gameday')
    
    ax.bar(map(lambda x: x + width, range(len(no_game))),
           no_game['avg(duration)'], width, color='k',
           alpha=0.4, label='Non-Gameday')

    labels = map(lambda x: x[:5], game['interval'])
    ax.set_xticks(range(41))
    ax.set_xticklabels(labels, rotation=90)
    ax.set_ylabel('Trip Duration (Secs)', fontsize=16)
    ax.tick_params(axis='both', labelsize=14)
    ax.set_title("Bus {}".format(bus), fontsize=20)
    ax.legend(fontsize=14)
```
